# **Data Scrapping (Easy Method)**

```python
from requests import get
from bs4 import BeautifulSoup
from warnings import warn
import numpy as np, pandas as pd
import requests
```

```python
#initialize empty lists to store the variables scraped
titles = []
counts = []
Critics_Consensus = []
synopsises = []
Starrings = []
directors = []

response = requests.get("https://editorial.rottentomatoes.com/guide/2021-best-movies")

#throw warning for status codes that are not 200
if response.status_code != 200:
    warn('Request: {}; Status code: {}'.format(requests, response.status_code))

#parse the content of current iteration of request
page_html = BeautifulSoup(response.text, 'html.parser')

movie_containers = page_html.find_all('div', class_ = 'row countdown-item')
```

```python
# Iterate over every movie and extract movies info and store them in variables
for container in movie_containers:

    title = container.find('div', class_ = 'article_movie_title').text
    titles.append(title)

    count = container.find('div', class_ = 'countdown-index-resposive').text
    counts.append(count)

    critics = container.find('div', class_ = 'info critics-consensus').text
    Critics_Consensus.append(critics)

    synopsis = container.find('div', class_ = 'info synopsis').text
    synopsises.append(synopsis)

    Starring = container.find('div', class_ = 'info cast').text
    Starrings.append(Starring)

    director = container.find('div', class_ = 'info director').text
    directors.append(director)


```

```python
# Define Columns and create the dataframe
a = {'movie': titles, 'count': counts, 'Critics Consensus': Critics_Consensus, 'synopsis': synopsises, 'Starring': Starrings, 'director': directors}
df = pd.DataFrame.from_dict(a, orient='index')
df = df.transpose()
```

```python
df.head()
```

```python
# Data Processing
df['movie'] = df['movie'].str.replace('\n', '')
df['Critics Consensus'] = df['Critics Consensus'].str.replace('Critics Consensus:', '')
df['synopsis'] = df['synopsis'].str.replace('Synopsis:', '')
df['Starring'] = df['Starring'].str.replace('\nStarring:', '')
df['director'] = df['director'].str.replace('\nDirected By:', '')
```

# **Table Scrapping from Wikipedia**


```python
import pandas as pd # library for data analysis
import requests # library to handle requests
from bs4 import BeautifulSoup # library to parse HTML documents

# get the response in the form of html
url = 'https://en.wikipedia.org/wiki/List_of_United_States_cities_by_area'
response=requests.get(url)
print(response.status_code)

# parse data from the html into a beautifulsoup object
soup = BeautifulSoup(response.text, 'html.parser')
indiatable=soup.find('table',{'class':"wikitable"})
```

```python
df=pd.read_html(str(indiatable))
# convert list to dataframe
df=pd.DataFrame(df[0])
df
```

```python
# Remove the first columns' names
df.columns=[multicols[-1] for multicols in df.columns]

# Rename the columns of the dataframe
Cols = ["Rank", "City", "State", "Land area (sq mi)", "Land area (km2)", "Water area (sq mi)", "Water area (km2)", "Total area (sq mi)", "Total area (km2)", "Population (2020)" ]
df.columns = Cols
df = df.drop('Rank', axis=1)
df
```

# **Data Scrapping (Many Pages)**

```python
from requests import get
from bs4 import BeautifulSoup
from warnings import warn
fimport numpy as np, pandas as pd
```

```python
# Note this takes about 40 min to run if np.arange is set to 9951 as the stopping point.

pages = np.arange(1, 50) # Last time I tried, I could only go to 10000 items because after that the URI has no discernable pattern to combat webcrawlers; I just did 4 pages for demonstration purposes. You can increase this for your own projects.
headers = {'Accept-Language': 'en-US,en;q=0.8'} # If this is not specified, the default language is Mandarin

#initialize empty lists to store the variables scraped
titles = []
years = []
ratings = []
genres = []
runtimes = []
imdb_ratings = []
imdb_ratings_standardized = []
metascores = []
votes = []

for page in pages:

   #get request for sci-fi
   response = get("https://www.imdb.com/search/title?genres=sci-fi&"
                  + "start="
                  + str(page)
                  + "&explore=title_type,genres&ref_=adv_prv", headers=headers)


   #throw warning for status codes that are not 200
   if response.status_code != 200:
       warn('Request: {}; Status code: {}'.format(requests, response.status_code))

   #parse the content of current iteration of request
   page_html = BeautifulSoup(response.text, 'html.parser')

   movie_containers = page_html.find_all('div', class_ = 'lister-item mode-advanced')

   #extract the 50 movies for that page
   for container in movie_containers:

       #conditional for all with metascore
       if container.find('div', class_ = 'ratings-metascore') is not None:

           #title
           title = container.h3.a.text
           titles.append(title)

           if container.h3.find('span', class_= 'lister-item-year text-muted unbold') is not None:

             #year released
             year = container.h3.find('span', class_= 'lister-item-year text-muted unbold').text # remove the parentheses around the year and make it an integer
             years.append(year)

           else:
             years.append(None) # each of the additional if clauses are to handle type None data, replacing it with an empty string so the arrays are of the same length at the end of the scraping

           if container.p.find('span', class_ = 'certificate') is not None:

             #rating
             rating = container.p.find('span', class_= 'certificate').text
             ratings.append(rating)

           else:
             ratings.append("")

           if container.p.find('span', class_ = 'genre') is not None:

             #genre
             genre = container.p.find('span', class_ = 'genre').text.replace("\n", "").rstrip().split(',') # remove the whitespace character, strip, and split to create an array of genres
             genres.append(genre)

           else:
             genres.append("")

           if container.p.find('span', class_ = 'runtime') is not None:

             #runtime
             time = int(container.p.find('span', class_ = 'runtime').text.replace(" min", "")) # remove the minute word from the runtime and make it an integer
             runtimes.append(time)

           else:
             runtimes.append(None)

           if float(container.strong.text) is not None:

             #IMDB ratings
             imdb = float(container.strong.text) # non-standardized variable
             imdb_ratings.append(imdb)

           else:
             imdb_ratings.append(None)

           if container.find('span', class_ = 'metascore').text is not None:

             #Metascore
             m_score = int(container.find('span', class_ = 'metascore').text) # make it an integer
             metascores.append(m_score)

           else:
             metascores.append(None)

           if container.find('span', attrs = {'name':'nv'})['data-value'] is not None:

             #Number of votes
             vote = int(container.find('span', attrs = {'name':'nv'})['data-value'])
             votes.append(vote)

           else:
               votes.append(None)
```

```python
sci_fi_df = pd.DataFrame({'movie': titles,
                      'year': years,
                      'rating': ratings,
                      'genre': genres,
                      'runtime_min': runtimes,
                      'imdb': imdb_ratings,
                      'metascore': metascores,
                      'votes': votes}
                      )

sci_fi_df.loc[:, 'year'] = sci_fi_df['year'].str[-5:-1] # two more data transformations after scraping
# Drop 'ovie' bug
# Make year an int
sci_fi_df['n_imdb'] = sci_fi_df['imdb'] * 10
final_df = sci_fi_df.loc[sci_fi_df['year'] != 'ovie'] # One small issue with the scrape on these two movies so just dropping those ones.
final_df.loc[:, 'year'] = pd.to_numeric(final_df['year'])
```
